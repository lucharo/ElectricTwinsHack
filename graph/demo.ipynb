```python
from plotly_functions import load_networkx_graph, select_subgraph_with_single_node, plot_subgraph_in_plotly, plot_profile_traffic, extract_data_with_query
```

```python
import pandas as pd
import pickle
```

```python
extra_data = pd.read_parquet("translated_posts.parquet")
profile_actvitity_link = extract_data_with_query("SELECT * FROM ProfileActivity")
combined_data = pd.merge(extra_data, profile_actvitity_link[["profile_id", "activity_id"]], left_on='id', right_on='activity_id', how='left')
```

```python
# graph = load_networkx_graph("subgraph_with_predictions.graphml")
with open('subgraph_with_predictions.pickle', 'rb') as f:
    graph = pickle.load(f)
```

```python
# overall_fig = plot_subgraph_in_plotly(graph)
```

```python
traffic_likelihood = combined_data.groupby("profile_id").sum()[["traffic_likelihood"]]
```

```python
# we can zoom into a suspicious individual based on how likely they are going to traffic
traffic_likelihood.reset_index().sort_values(by="traffic_likelihood", ascending=False)
```

```python
profile_id_to_explore = 10558
```

```python
subgraph = select_subgraph_with_single_node(graph, profile_id_to_explore)
```

```python
plot_subgraph_in_plotly(subgraph)
```

```python
plot_profile_traffic(profile_id_to_explore)
```

```python
# suspicious people connected to this individual by graph on top of the llm prediction
neighbors = set()
neighbors.update(graph.neighbors(profile_id_to_explore))
```

```python
for node in neighbors:
    graph_based_prediction = graph.nodes[node]["graph_based_prediction"]
    if graph_based_prediction == "suspicious":
        print(node, "graph based prediction",graph.nodes[node]["graph_based_prediction"])
        print(node, "llm based prediction", graph.nodes[node]["llm_based_prediction"])
```
